```python
import matplotlib.pyplot as plt
sns.lineplot(x="sepal_length",y="sepal_width",data=data)
plt.title("Title using Matplot Function")
plt.show()
```

```python
sns.barplot(x="species",y="sepal_width",data=data)
plt.title("Barplot for species")
plt.show()
```

```python
sns.countplot(x="species",data=data)
plt.title("Countplot for species")
plt.show()
```

```python
sns.countplot(x="petal_length",data=data)
plt.title("Countplot for petal length")
plt.show()
```
