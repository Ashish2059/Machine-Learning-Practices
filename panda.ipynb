```python
import  pandas as pd
import numpy as np
#create simple array
data=np.array(['P','Y','T','H','O','N'])
print(data)
```

```python
s=pd.Series(data)
print(s)
```

```python
#geterate series with data and index
id=[1,3,4,5,6,2,9]
Data=['a','b','c','d','e','f','g']
a=pd.Series(Data,index=id)
print(a)
```

```python
#generate the series with a dictionary data
dct={'red':1,'blue':2}
a=pd.Series(dct)
print(a)
```

```python
#creating data frame
data={'Name':['Tom','Nick','Krish'],
      'Age':[20,21,45]}
df=pd.DataFrame(data)
print(df)
```

```python
df=pd.DataFrame(({'c1':[2,3],'c2':[4,5]}))
print(df.size)
```

```python
d={'one':pd.Series([1,2,3,],index=['a','b','c']),
   'two':pd.Series([1,2,3,4,],index=['a','b','c','d'])}
df=pd.DataFrame(d)
print(df)
print("Adding New Column:")
df['three']=pd.Series([10,20,30],index=['a','b','c'])
print(df)
```
